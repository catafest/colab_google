This is not a complete course on deep learning is just one tutorial about **fashion_mnist** dataset and show the content of this dataset.

Let's install all python modules.


```python
import tensorflow as tf
import tensorflow_datasets as tfds
from tensorflow import keras
```

Import the **fashion_mnist** dataset like *fashion_data*

```python

fashion_data = keras.datasets.fashion_mnist.load_data()
(x_train,y_train),(x_val,y_val)= fashion_data
```

Show the size of *x_train.shape* and *y_train.shape*:

```python
x_train.shape, y_train.shape
```

You can see this data set comes with 60,000 samples in our training set, and the images are 28 pixels x 28 pixels each.

Show the size of *x_val.shape* and *y_val.shape*:

```python
x_val.shape ,y_val.shape
```

The next source code will show the first 3 images with the shape of the dataset:


```python
import matplotlib.pyplot as plt
for i in range(0, 3):
  plt.imshow(x_train[i])
  plt.colorbar()
  plt.show()
```

```python

plt.figure(figsize=(7,7))
for i in range(1,65):
  plt.subplot(8,8,i)
  plt.imshow(x_train[i])
  plt.title(y_train[i],color='r')
  plt.xticks([])
  plt.yticks([])
plt.tight_layout()
plt.show()
```

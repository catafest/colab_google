```python
from IPython.display import Image
from diffusers import StableDiffusionPipeline
import torch

model_id = "runwayml/stable-diffusion-v1-5"
pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16)
pipe = pipe.to("cuda")

prompt = "An image of a girl into comics style"
image_001 = pipe(prompt).images[0]  
filename_001 = "test_001.png"
image_001.save(filename_001)
Image(filename_001)
```

```python
from IPython.display import Image
from diffusers import DiffusionPipeline

pipeline = DiffusionPipeline.from_pretrained("runwayml/stable-diffusion-v1-5")
pipeline.to("cuda")
image_002 = pipeline("An image of a girl into comics style").images[0]
filename_002 = "test_002.png"
image_002.save(filename_002)
Image(filename_002)
```
